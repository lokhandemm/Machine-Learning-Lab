## Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
#Read the dataset
df=pd.read_csv('Social_Network_Ads.csv')
print(df.head())
```

## Splitting the dataset into the Training set and Test set

```python
X = df.drop(['Purchased'], axis=1)  # Features
y = df['Purchased']  # Labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Feature Scaling

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Display the first few rows of the dataset
print("Original Data:")
print(df.head())

# Apply feature scaling to the 'Age' and 'EstimatedSalary' columns
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Age', 'EstimatedSalary']])

# Create a DataFrame with the scaled features
scaled_data = pd.DataFrame(scaled_features, columns=['Age_Scaled', 'EstimatedSalary_Scaled'])

# Combine the scaled data with the original dataset
data_scaled = pd.concat([df, scaled_data], axis=1)

# Display the first few rows of the scaled dataset
print("\nScaled Data:")
print(data_scaled.head())
```

## Training the Decision Tree Classification model on the Training set

```python
#model = DecisionTreeClassifier(random_state=42)
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X_train, y_train)
```

# Making Predictions:

```python
y_pred = model.predict(X_test)
#predict(): The trained model predicts the species for the test data (X_test).
```

# Evaluating the Model:

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy:", accuracy)
print("Classification Report:\n", report)
#accuracy_score(): Calculates the accuracy of the model (how many predictions were correct out of total predictions).
#classification_report(): Provides a detailed report of precision, recall, and F1-score for each class.
```

# Visualizing the Decision Tree:

```python
# Define feature names (replace with actual feature names from your dataset)
feature_names = X.columns.tolist()

# Define class names (replace with actual class names for your target variable)
class_names = ['No', 'Yes']  # Adjust based on your target variable mapping

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(model, filled=True, feature_names=feature_names, class_names=class_names)
plt.show()
```

The decision tree model provides a clear visual representation of how age and estimated salary influence purchasing behavior, with each node representing a decision based on these features. The model effectively splits the data to classify users into those who are likely or unlikely to make a purchase.
